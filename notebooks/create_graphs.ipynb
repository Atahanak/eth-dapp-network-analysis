## Print Edge List

```python
name = "Ethex"
with open(graph_path + name + "/" + name + '.json') as f:
    graph_info = json.load(f)
graph_info['name'], graph_info['#nodes'], graph_info['#edges']

edges = graph_info["edges"]
out = graph_path + name + "/" + name + ".edgelist"

labels = {}
c = 0

start = time.time()
print("Creating graph", out, "...", end='', flush=True)
#get smallest time stamp
mint = sys.maxsize
for e in edges:
        t = int(e[2])
        if t < mint:
                mint = t
with open(out, "w") as o:
        for e in edges:
                u = e[0]
                v = e[1]
                t = e[2]
                if u not in labels:
                        labels[u] = c
                        c += 1
                u = labels[u]
                if v not in labels:
                        labels[v] = c
                        c += 1
                v = labels[v]
                o.write(str(u) + " " + str(v) + " " + str(int(t) - mint) + "\n")
end = time.time()
print("took", end - start, "seconds.")
```
