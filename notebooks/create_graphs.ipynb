## Print Edge List

```python
edges = graph["edges"]
out = name + "/" + name + ".edgelist"

labels = {}
c = 0

start = time.time()
print("Creating graph", out, "...", end='', flush=True)
#get smallest time stamp
mint = sys.maxsize
for e in edges:
        t = int(e[2])
        if t < mint:
                mint = t
with open(out, "w") as o:
        for e in edges:
                u = e[0]
                v = e[1]
                t = e[2]
                if u not in labels:
                        labels[u] = c
                        c += 1
                u = labels[u]
                if v not in labels:
                        labels[v] = c
                        c += 1
                v = labels[v]
                o.write(str(u) + " " + str(v) + " " + str(int(t) - mint) + "\n")
end = time.time()
print("took", end - start, "seconds.")
```
